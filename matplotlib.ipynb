```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
x=[0,1,2,3,4,5]
y=[0,2,4,6,8,10]

#plt.plot(x,y,label="2x",color="blue",linewidth=2,marker=".",linestyle="--",markersize=10,markeredgecolor="red")
plt.plot(x,y,"r.--",label="2x")
x2=np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,"b",label="x^2")

plt.plot(x2[5:],x2[5:]**2,"g--",label="x^2")
plt.title("our first plot",fontdict={"fontname":"Comic Sans Ms"})
plt.xlabel("x axis",fontdict={"fontname":"Comic Sans Ms"})
plt.ylabel("y axis",fontdict={"fontname":"arial"})

plt.legend()
plt.xticks([0,1,2,3,4,5])
#plt.yticks([2,3,4,5,6])
#plt.savefig("first_graph.png",dpi=300)
plt.show()
```
