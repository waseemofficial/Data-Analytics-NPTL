# <center> <span style="color:lightblue"> 3D Plot</span>

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.animation import PillowWriter
%matplotlib inline
```

```python
import random
x=[]
y=[]
z=[]
for i in range(20):
    x.append(random.randint(1,15))
    y.append(random.randint(1,25))
    z.append(random.randint(5,35))
```

```python
ax=plt.axes(projection="3d")
ax.scatter(x,y,z)
plt.show()
```

```python
x=np.random.randint(0,20,(20,))
y=np.random.randint(0,15,(20,))
z=np.random.randint(5,35,(20,))
ax=plt.axes(projection="3d")
ax.scatter(x,y,z,marker="v")
plt.show()
```

```python
ax=plt.axes(projection="3d")
x=np.arange(0,50,0.1)
y=np.arange(0,50,0.1)
z=x*y
ax.plot(x,y,z)
plt.show()
```

```python
ax=plt.axes(projection="3d")
x=np.arange(0,50,0.1)
y=np.arange(0,50,0.1)
z=np.sign(x)*np.cos(y)
ax.set_title("3D Plot")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.plot(x,y,z)
plt.show()
```

```python
ax=plt.axes(projection="3d")
x=np.arange(0,50,0.1)
y=np.arange(0,50,0.1)
X,Y=np.meshgrid(x,y)
Z=X*Y
ax.plot_surface(X,Y,Z)
plt.show()
```

```python
ax=plt.axes(projection="3d")
x=np.arange(-5,5,0.1)
y=np.arange(-5,5,0.1)
X,Y=np.meshgrid(x,y)
Z=np.sin(X)*np.cos(Y)
ax.plot_surface(X,Y,Z,cmap="plasma")
plt.show()
```
