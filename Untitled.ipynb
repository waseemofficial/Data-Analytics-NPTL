```python
import pandas as pd
```

```python
import requests
from bs4 import BeautifulSoup

url= "https://prod--api.educationdatahub.dsti.gov.sl/schools/2019"
r =requests.get(url)
soup=BeautifulSoup(r.content)

p=soup.find_all("p")
```

```python
data = r.json()
```

```python
school = []

for id in data:
    school.append(id)
```

```python
df = pd.json_normalize(school)
```
