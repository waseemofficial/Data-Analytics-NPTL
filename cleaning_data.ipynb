# <center> Data Cleaning using python

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly as pl
#import folum as fo

%matplotlib inline
```

```python
#import Data
data= pd.read_excel(r"Customer Call List.xlsx")
#data.head(10)
```

```python
#Drop duplicate data
data = data.drop_duplicates()
data.head()
```

```python
#Drop columns 
data = data.drop(columns = "Not_Useful_Column")
data.head()
```

```python

data["Last_Name"] = data["Last_Name"].str.strip("123._/")
data.head()
```

```python
data["Phone_Number"] = data["Phone_Number"].str.replace('nan--','')

data["Phone_Number"] = data["Phone_Number"].str.replace('Na--','')
data.head()
```

```python
data[["Street_Address", "State", "Zip_Code"]] = data["Address"].str.split(',',2, expand=True)
data
```

```python
data["Do_Not_Contact"] = data["Do_Not_Contact"].str.replace('Yes','Y')

data["Do_Not_Contact"] = data["Do_Not_Contact"].str.replace('No','N')
data.head()
```

```python
#data = data.replace('N/a','')
#data = data.replace('NaN','')


data=data.fillna('')
data.head()
```

```python
for x in data.index:
    if data.loc[x, "Do_Not_Contact"] == 'Y':
        data.drop(x, inplace=True)

data.head()
```

```python
for x in data.index:
    if data.loc[x, "Phone_Number"] == '':
        data.drop(x, inplace=True)

data.head()

#Another way to drop null values
#data = data.dropna(subset="Phone_Number"), inplace=True)
```

```python
#Reset Index after removing rows
data = data.reset_index(drop=True)
data.head()
```

```python
data["Phone_Number"]=data["Phone_Number"].apply(lambda x: str(x))
data["Phone_Number"].apply(lambda x: x[0:3]+ "-" + x[3:6] + "-" x[6:10])
data["Phone_Number"]=data["Phone_Number"].apply(lambda x: str(x))
data["Phone_Number"]=data["Phone_Number"].apply(lambda x: x[0:3] + "-" + x[3:6] + "-" + x[6:10])
data["Phone_Number"] = data["Phone_Number"].str.replace("[^a-zA-Z0-9]","")
```
